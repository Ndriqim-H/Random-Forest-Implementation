```python
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np


def Random_Forest_Fit(Training_Data, n_features, n_samples, n_estimators=100, max_depth=None, target = 'target'):
    trees = []
    for i in range(n_estimators):
        sample = resample(Training_Data, replace=True, n_samples=n_samples)
        X_train = sample.drop(target, axis=1)
        Y_train = sample[target]
        tree = DecisionTreeClassifier(max_depth=max_depth, max_features=n_features)
        features = resample(X_train.columns, replace=False, n_samples=n_features)
        X_train = X_train[features]
        tree.fit(X_train, Y_train)
        trees.append(tree)
    return trees

# def Random_Forest_Predict(X, trees):
#     predictions = np.zeros(len(X))
#     for tree in trees:
#         predictions += tree.predict(X)
#     return np.round(predictions / len(trees))
    
def Random_Forest_Predict(X, trees):
    predictions = np.zeros((len(X), len(trees)))
    for i, tree in enumerate(trees):
        features = tree.tree_.feature
        X_test = X[:, features]
        predictions[:, i] = tree.predict(X_test)
    final_predictions = []
    for i in range(len(X)):
        counts = np.bincount(predictions[i].astype(int))
        final_predictions.append(np.argmax(counts))
    return np.array(final_predictions)

# def Find_Best_Feature(features, X, Y):
#     best_feature = None

#     return best_feature
```
